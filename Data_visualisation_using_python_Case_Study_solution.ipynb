## Dataset Description
<br>

**battery_power** : Total energy a battery can store in one time measured in mAh

**pc** : Primary Camera mega pixels

**px_width** : Pixel Resolution Width

**ram** : Random Access Memory in Megabytes

**sc_h** : Screen Height of mobile in cm

**sc_w** : Screen Width of mobile in cm

**talk_time** : longest time that a single battery charge will last.

**three_g** : Mobile phone has 3G or not

**touch_screen** : Mobile phone has touch screen or not

**wifi** : Mobile phone has wifi or not

**price_range** : Price Range of the mobile phone

### <a id='toc1_5_3_'></a>[Pie chart](#toc0_)
Pie chart visualization 

* We're using Plotly for the pie chart.
* `pie()` function is used for plotting the pie chart, which contains `names` `labels` andÂ Â `color sequence` as arguments.
*Â `update_traces()`Â functionÂ willÂ updateÂ theÂ positionÂ ofÂ theÂ labelsÂ andÂ textinfo.Â <br>



```python
fig = px.pie(df, names='m_dep',labels={'m_dep':'m depart'},color_discrete_sequence=px.colors.sequential.RdBu) #px.pie
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
fig = px.pie(df, names='pc',labels={'pc':'pc'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Here, we observe that most of the mobiles have a primary camera with 10 megapixels.

### <a id='toc1_5_4_'></a>[Violin Plot](#toc0_)

A Violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.



```python
sns.violinplot(data=df, x="three_g",y="four_g")
```

This means that all 4G phones also contain 3G technology, but not the other way around.

```python
# plotting the violin plot
fig = px.violin(df, y="int_memory", x="price_range")
 
# showing the plot
fig.show()
```

From the violin plot, we can understand the distribution of the internal memory in the various mobiles in their price ranges.

* In categories 0 and 1, we can see a higher density toward low internal memory.
* In category 3, we can say that higher density leads to greater internal memory.

### <a id='toc1_5_6_'></a>[Joint Plot](#toc0_)


Jointplot provides a convenient interface to the class of joint grid by using several types of points. It is intended for a lightweight wrapper; suppose we need more flexibility, we are using a joint grid directly. We draw a plot by using two variables to define the seaborn jointplot. It is very important; also, we are drawing two graphs using univariate and bivariate.



```python
# Joint kernel density estimate
g = sns.jointplot(
    data=df,
    x="price_range", y="ram", hue="three_g",
    kind="kde",
)
```

This graph illustrates the relationship between RAM and price range in relation to 3G connectivity.

* This indicates that 3G is extensively utilized in RAM and price levels.

```python
h = sns.jointplot(
    data=df,
    x="three_g", y="four_g",
    kind="kde",
)
```

From this graph, we can say every mobile device that has 4G has 3G in it. However, it is not the opposite.Â 

### <a id='toc1_5_7_'></a>[Scatter Plot](#toc0_)


A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axes indicates values for an individual data point. Scatter plots are used to observe relationships between variables.




```python
plt.figure(figsize = (10, 8))
sns.scatterplot(data=df, x="battery_power", y="px_height", hue="four_g")
```

When we compare the pexel height and battery power of 4G and non-4G mobile devices, we can see that the distributions of both are very similar.

### <a id='toc1_5_8_'></a>[Box plot](#toc0_)

Box plotsÂ (also calledÂ box-and-whisker plotsÂ orÂ box-whisker plots) give a good graphical image of the concentration of the data. They also show how far the extreme values are from most of the data. A box plot is constructed from five values: the minimum value, the first quartile, the median, the third quartile, and the maximum value.



```python
plt.figure(figsize = (15, 6))
sns.boxplot(x='price_range', y='battery_power', data=df)
sns.set_style("dark")
plt.show()
```

As a result, we can conclude that as the price rises, so does the battery power.Â 

```python
fig = px.box(df, x="talk_time", y="battery_power")
fig.show()
```

We can conclude that a battery capacity of 1000 to 1400 mAh is adequate for regular talk time.Â

### <a id='toc1_5_10_'></a>[Hexagonal Binning plot](#toc0_)

```python
X=df["pc"]
Y=df["fc"]
# changing the rc parameters and plotting
plt.rcParams['figure.figsize'] = [15, 8]

# Plotting the hexbin plot
fig, ax = plt.subplots()
ax.hexbin(X, Y, gridsize=20)
#ax.set(xlim=(-1, 20), ylim=(-1, 20))  limiting the x and y axis for better visualization 
plt.show()
```

This indicates that both the primary camera and the face camera are widely spread.

### <a id='toc1_5_9_'></a>[Kde Plot](#toc0_)

Kernel Distribution Estimation Plot which displays the probability density function of continuous or non-parametric data variables; thus, we are able to plot for both univariate and multivariate data sets.



```python
sns.kdeplot(data=df, x="battery_power")
```

Here, we can observe that the battery capacity ranges greatly from more than 500 to less than 2000.

```python
sns.kdeplot(data=df, x="n_cores")
```

Here, we can observe that the number of cores ranges from 1 to 9.

```python
sns.kdeplot(data=df, x="talk_time", hue="four_g", multiple="stack")
```

On devices without 4G, the talk time density is higher.

```python
plt.figure(figsize = (10, 10))
sns.kdeplot(data=df, x="n_cores", y="price_range")
```

From the following, we can say :
* In the price range of 0 to 1, there is a high density on 2 to 4 cores.<br>
* There is a high density on 4 to 8 cores in the 2 to 3 price range.Â 

```python
sns.kdeplot( data=df, x="battery_power", y="mobile_wt", hue="price_range", fill=True)
```

This plot shows that devices in price range 3 have a battery capacity of 1500-2000 mAH, while the rest are evenly distributed.Â
